```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
with open("/content/data.txt","r") as file:
  text = file.read()
print(text)
```

```python
tokens = word_tokenize(text)
print(tokens)
```

```python

pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

```python
tokens_new = [word for word in tokens if word not in string.punctuation]
print(tokens_new)
```

```python
stop_words = set(stopwords.words('english'))
filtered_token = [word for word in tokens_new if word.lower() not in stop_words]

print(filtered_token)
```

```python
porter = PorterStemmer()

stemmed_words = [porter.stem(word) for word in filtered_token]
print(stemmed_words)
```

```python

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_token]

print(lemmatized_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Transform documents into TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform([res , res2])
#print(tfidf_matrix)
# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()
#print(feature_names)

print(tfidf_matrix)
# Convert matrix to array and print values
for i, doc in enumerate(tfidf_matrix.toarray()):
    print(f"Document {i+1}:")
    for word, score in zip(feature_names, doc):
        if score > 0:
            print(f"  {word}: {score:.4f}")
```
